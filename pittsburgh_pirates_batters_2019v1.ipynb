```python
may_games = 21
may_ab = may_totals[1]
may_runs = round(may_totals[2])
may_hits = round(may_totals[3])
may_2b = round(may_totals[4])
may_3b = round(may_totals[5])
may_hr = round(may_totals[6])
may_rbi = round(may_totals[7])
may_bb = round(may_totals[8])
may_so = round(may_totals[9])
may_sb = round(may_totals[10])
may_cs = round(may_totals[11])
may_hbp = may_totals[12]
may_sf = may_totals[13]
may_tb = may_totals[14]
```

```python
may_avg = round(may_hits/may_ab, 3)
may_obp = round(((may_hits + may_bb + may_hbp)/(may_ab + may_bb + may_hbp + may_sf)), 3)
may_slg = round((may_tb/may_ab), 3)
```

```python
may_ops = round((may_obp + may_slg), 3)
```

```python
total_games = april_games + may_games
total_runs = april_runs + may_runs
total_hits = april_hits + may_hits
total_2b = april_2b + may_2b
total_3b = april_3b + may_3b
total_hr = april_hr + may_hr
total_rbi = april_rbi + may_rbi
total_bb = april_bb + may_bb
total_so = april_so + may_so
total_sb = april_sb + may_sb
total_cs = april_cs + may_cs
```

```python
total_avg = round(((total_hits/(april_ab + may_ab))), 3)
total_obp = round(((april_obp + may_obp)/2), 3)
total_slg = (april_slg + may_slg)/2
total_ops = round(((april_ops + may_ops)/2), 3)
```
